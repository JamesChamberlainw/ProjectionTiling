# Select a Projection and Input Polygon 

This is just a basic example showing the most basic useage of this to convert a polygon to a set of projection aligned tiles and uploading it to GEE assets. 

For a more in-depth breakdown, go to the interactive example file. You don't need to use the geemap example, but the code will work without the map so long as you provide an input polygon and projection. The only requirements for this to work are the Earth Engine API (`ee`) and default Python environment libraries `time`, `math`, and `re`.

```python
# Example Polygon (Exeter, UK)
polygon = ee.Geometry.Polygon([[[-3.548069, 50.735695], [-3.566608, 50.735586], [-3.563862, 50.708961], [-3.514938, 50.676229], [-3.483009, 50.670789], [-3.442154, 50.678839], [-3.394089, 50.733413], [-3.427734, 50.74797], [-3.488503, 50.753618], [-3.548584, 50.748187], [-3.548069, 50.735695]]])

# Select Projection (any)
projection = ee.ImageCollection("COPERNICUS/S2_SR_HARMONIZED") \
    .filterBounds(polygon) \
    .first() \
    .select("B1") \
    .projection()

# Change the assetId_prefix to your own username and desired folder in your EE assets. - will result in asset_name_tile_1, to ..._tile_n
# This will create a set of shapefiles over the input polygon with 1 pixel width/height in the selected projection. (B1 = 60m per pixel)
# all major parts to this tool are replaceable (e.g., output format can be changed by uploading your own function (see tiler_interactive_example.ipynb))
# the size of the output shapefiles can be changed via adjusting shpfile_max_polygons - deafult is (71**2) 5041 what is known to wokr but larger and smaller shape files can be created ideally 
# the identifier column can be set via identifier_col (by default that is jsut "identifier") 
#       and is set by the input lambda fn identifier_name0_fn (seen in default inputs to projTiler)
projTiler(polygon, projection, width_px=1, shpfile_max_polygons=71**2, assetId_prefix="users/jameswilliamchamberlain/tiler_example_basics")


# Additional Notes: 
# projTiler is a class that can be stored and if you are having issues the class stores all inputs and outputs 
#    so you check the __init__ for any of those but...
#  the main output is stored under tile_lst what is a lst of ee.FeatureCollection objects. So that more than likely what you will want
# tiler = projTiler(polygon, projection, width_px=6, export_fn=None ... etc) # Note you do not need to set a output but it would result in no output 
# tile_list = tiler.tile_lst  # extract to local variable (note this is a list and its up to you from here if you want to work with it this way)
```
